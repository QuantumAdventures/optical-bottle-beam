```python
def prob_distr(data, C, k1, k2, k3):
    kb = 1.380649*(10**-23)
    T = 293
    r, z = data
    beta = 1/(kb*T)
    U = k1*np.power(r, 4) + k2*np.power(r,2)*np.power(z,2)+k3*np.power(z,4)
#    U = k1*np.power(r, 4) +k3*np.power(r,2)*np.power(z,2)+k5*np.power(z,4)


    return C*r*np.exp(-beta*U)

def create_3d_data(r, z):

    df = pd.DataFrame()
    df['r'] = r
    df['z'] = z
    def __convert_interval_to_float__(interval):
        splitted = str(interval)[1:-1].split(',')
        return 0.5*(float(splitted[0])+float(splitted[1]))
    rs, zs, counts = [], [], []
    groups = df.groupby([pd.cut(df.r, np.linspace(0, 1.2*10**-7, 15)),pd.cut(df.z, np.linspace(-1.5*10**-7, 1.5*10**-7, 15))])
    grouped_df = groups.size().unstack()
    for i, row in grouped_df.iterrows():
        for index_i, index in enumerate(row.index):
            rs.append(convert_interval_to_float(row.name))
            zs.append(convert_interval_to_float(index))
            counts.append(row.iloc[index_i])
    rs = np.array(rs)
    zs = np.array(zs)
    counts = np.array(counts)
    return rs, zs, counts
```

```python
rs, zs, counts = create_3d_data(r, z)



limits = np.array([10**12, 10**10, 10**10, 10**10])
params, _ = curve_fit(prob_distr, (rs, zs), counts, bounds=(-limits, limits))


fig = plt.figure(figsize=plt.figaspect(0.5))

ax = fig.add_subplot(1, 2, 1, projection='3d')
X1, Y1 = np.meshgrid(rs, zs)
Z1 = prob_distr((X1, Y1), params[0], params[1], params[2], params[3])
ax.plot_wireframe(X1, Y1, Z1)

#[ 1.75875910e+11  8.77948988e+08 -5.93812455e+07 -2.17289712e+06
# -5.58542864e-09  3.03523185e+07]

ax = fig.add_subplot(1, 2, 2, projection='3d')
hist, xedges, yedges = np.histogram2d(r, z, bins=15, range=[[0, 1.2*10**-7], [-1.5*10**-7, 1.5*10**-7]])
# Construct arrays for the anchor positions of the 16 bars.
xpos, ypos = np.meshgrid(xedges[:-1], yedges[:-1], indexing="ij")
#xpos = xpos.ravel()
#ypos = ypos.ravel()
zpos = 0
# Construct arrays with the dimensions for the 16 bars.
#dx = dy = 0.15 * 10**-7* np.ones_like(zpos)
#dz = hist.ravel()
#ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')
ax.plot_wireframe(xpos, ypos, hist)
plt.show()
```

```python
def prob_distr_gaussian(z, kz, C):
    kb = 1.380649*(10**-23)
    T = 293
    beta = 1/(kb*T)
    return np.sqrt(2*np.pi/beta)*C*np.exp(-beta*kz*np.power(z, 2)/2)
```
